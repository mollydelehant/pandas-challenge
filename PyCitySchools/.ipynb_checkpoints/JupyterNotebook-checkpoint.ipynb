```python
import pandas as pd

# Load the data files
school_data_to_load = ("C:/Users/tpw3615/OneDrive - Northwestern University/Desktop/pandas-challenge/Starter_Code/PyCitySchools/Resources/schools_complete.csv")
student_data_to_load = ("C:/Users/tpw3615/OneDrive - Northwestern University/Desktop/pandas-challenge/Starter_Code/PyCitySchools/Resources/students_complete.csv")

school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

# Calculate the total number of unique schools
total_schools = school_data_complete["school_name"].nunique()

# Calculate the total number of students
total_students = school_data_complete["Student ID"].count()

# Calculate the total budget
total_budget = school_data["budget"].sum()

# Calculate the average math score
average_math_score = school_data_complete["math_score"].mean()

# Calculate the average reading score
average_reading_score = school_data_complete["reading_score"].mean()

# Calculate the percentage of students who passed math
passing_math = school_data_complete[school_data_complete["math_score"] >= 70]
percentage_passing_math = (passing_math["Student ID"].count() / total_students) * 100

# Calculate the percentage of students who passed reading
passing_reading = school_data_complete[school_data_complete["reading_score"] >= 70]
percentage_passing_reading = (passing_reading["Student ID"].count() / total_students) * 100

# Calculate the percentage of students who passed both math and reading
passing_math_reading = school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)]
percentage_passing_overall = (passing_math_reading["Student ID"].count() / total_students) * 100

# Create a DataFrame for the District Summary
district_summary = pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [average_math_score],
    "Average Reading Score": [average_reading_score],
    "% Passing Math": [percentage_passing_math],
    "% Passing Reading": [percentage_passing_reading],
    "% Overall Passing": [percentage_passing_overall]
})

# Display the district's key metrics
print("District Summary:")
print(district_summary)
```

```python
# Step 1: Group the student data by the school name
school_groups = student_data.groupby('school_name')

# Step 2: Calculate the total number of students in each school
total_students = school_groups['Student ID'].count()

# Step 3: Calculate the total school budget for each school
total_school_budget = school_data.set_index('school_name')['budget']

# Step 4: Calculate the per-student budget
per_student_budget = total_school_budget / total_students

# Step 5: Calculate the average math and reading scores for each school
average_math_score = school_groups['math_score'].mean()
average_reading_score = school_groups['reading_score'].mean()

# Step 6: Calculate the percentage of students passing math and reading in each school
passing_math = student_data[student_data['math_score'] >= 70].groupby('school_name')['Student ID'].count() / total_students * 100
passing_reading = student_data[student_data['reading_score'] >= 70].groupby('school_name')['Student ID'].count() / total_students * 100

# Step 7: Calculate the percentage of students passing both math and reading in each school
overall_passing = student_data[(student_data['math_score'] >= 70) & (student_data['reading_score'] >= 70)].groupby('school_name')['Student ID'].count() / total_students * 100

# Create a DataFrame to store the school summary
school_summary = pd.DataFrame({
    "School Type": school_data.set_index('school_name')['type'],
    "Total Students": total_students,
    "Total School Budget": total_school_budget,
    "Per Student Budget": per_student_budget,
    "Average Math Score": average_math_score,
    "Average Reading Score": average_reading_score,
    "% Passing Math": passing_math,
    "% Passing Reading": passing_reading,
    "% Overall Passing": overall_passing
})

# Format the columns
school_summary["Total School Budget"] = school_summary["Total School Budget"].map('${:,.2f}'.format)
school_summary["Per Student Budget"] = school_summary["Per Student Budget"].map('${:,.2f}'.format)

# Display the school summary DataFrame
school_summary

```

```python
# Sort the schools by % Overall Passing in descending order
top_schools = school_summary.sort_values(by="% Overall Passing", ascending=False)

# Display the top 5 rows
top_schools.head()
```

```python
# Group the data by school name and grade, and calculate the mean math score
math_scores_by_grade = student_data.groupby(['school_name', 'grade'])['math_score'].mean()

# Convert the Series to a DataFrame
math_scores_by_grade = math_scores_by_grade.reset_index()

# Pivot the DataFrame to have grades as columns
math_scores_by_grade = math_scores_by_grade.pivot(index='school_name', columns='grade', values='math_score')

# Reorder columns to match the desired order (9th, 10th, 11th, 12th)
math_scores_by_grade = math_scores_by_grade[['9th', '10th', '11th', '12th']]

# Rename columns for clarity
math_scores_by_grade.columns.name = None  # Remove column name
math_scores_by_grade = math_scores_by_grade.rename(columns={
    '9th': '9th Grade',
    '10th': '10th Grade',
    '11th': '11th Grade',
    '12th': '12th Grade'
})

# Display the DataFrame
math_scores_by_grade
```

```python
# Group the data by school name and grade, and calculate the mean reading score
reading_scores_by_grade = student_data.groupby(['school_name', 'grade'])['reading_score'].mean()

# Convert the Series to a DataFrame
reading_scores_by_grade = reading_scores_by_grade.reset_index()

# Pivot the DataFrame to have grades as columns
reading_scores_by_grade = reading_scores_by_grade.pivot(index='school_name', columns='grade', values='reading_score')

# Reorder columns to match the desired order (9th, 10th, 11th, 12th)
reading_scores_by_grade = reading_scores_by_grade[['9th', '10th', '11th', '12th']]

# Rename columns for clarity
reading_scores_by_grade.columns.name = None  # Remove column name
reading_scores_by_grade = reading_scores_by_grade.rename(columns={
    '9th': '9th Grade',
    '10th': '10th Grade',
    '11th': '11th Grade',
    '12th': '12th Grade'
})

# Display the DataFrame
reading_scores_by_grade
```

```python
# Create the spending bins and labels
spending_bins = [0, 585, 630, 645, 680]
spending_labels = ["<$585", "$585-630", "$630-645", "$645-680"]

# Categorize school spending based on the bins and labels
per_school_summary['Spending Ranges (Per Student)'] = pd.cut(per_school_summary['Per Student Budget'], spending_bins, labels=spending_labels)

# Calculate the mean math and reading scores per spending range
spending_math_scores = per_school_summary.groupby("Spending Ranges (Per Student)")["Average Math Score"].mean()
spending_reading_scores = per_school_summary.groupby("Spending Ranges (Per Student)")["Average Reading Score"].mean()

# Calculate the mean percentage of students passing math, passing reading, and overall passing per spending range
spending_passing_math = per_school_summary.groupby("Spending Ranges (Per Student)")["% Passing Math"].mean()
spending_passing_reading = per_school_summary.groupby("Spending Ranges (Per Student)")["% Passing Reading"].mean()
overall_passing_spending = per_school_summary.groupby("Spending Ranges (Per Student)")["% Overall Passing"].mean()

# Create the spending_summary DataFrame
spending_summary = pd.DataFrame({
    "Average Math Score": spending_math_scores,
    "Average Reading Score": spending_reading_scores,
    "% Passing Math": spending_passing_math,
    "% Passing Reading": spending_passing_reading,
    "% Overall Passing": overall_passing_spending
})

# Display the spending_summary DataFrame
spending_summary


```

```python
# Create bins and labels for school size
size_bins = [0, 1000, 2000, 5000]
size_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

# Categorize schools based on size bins
per_school_summary["School Size"] = pd.cut(per_school_summary["Total Students"], size_bins, labels=size_labels)

# Calculate the mean math and reading scores per school size
size_math_scores = per_school_summary.groupby("School Size")["Average Math Score"].mean()
size_reading_scores = per_school_summary.groupby("School Size")["Average Reading Score"].mean()

# Calculate the mean percentage of students passing math, passing reading, and overall passing per school size
size_passing_math = per_school_summary.groupby("School Size")["% Passing Math"].mean()
size_passing_reading = per_school_summary.groupby("School Size")["% Passing Reading"].mean()
size_overall_passing = per_school_summary.groupby("School Size")["% Overall Passing"].mean()

# Create the size_summary DataFrame
size_summary = pd.DataFrame({
    "Average Math Score": size_math_scores,
    "Average Reading Score": size_reading_scores,
    "% Passing Math": size_passing_math,
    "% Passing Reading": size_passing_reading,
    "% Overall Passing": size_overall_passing
})

# Display the size_summary DataFrame
size_summary

# Group the per_school_summary DataFrame by "School Type" and calculate the averages
type_summary = per_school_summary.groupby("School Type")[["Average Math Score", "Average Reading Score",
                                                       "% Passing Math", "% Passing Reading", "% Overall Passing"]].mean()

# Display the type_summary DataFrame
type_summary
```
